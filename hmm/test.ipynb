```python
import numpy as np
import pathlib
import matplotlib.pyplot as plt

#load in a .npy file
def load_npy_file(file_path):
    try:
        data = np.load(file_path, allow_pickle=True)
        return data
    except Exception as e:
        print(f"Error loading .npy file: {e}")
        return None

home_dir = pathlib.Path(r"C:\Users\Matthew\Documents\steve")

experiment = home_dir / 'sub-005_id-1122877_type-wtshelterswitch' #'sub-006_id-1123131_type-wtshelterswitch' #'sub-005_id-1122877_type-wtshelterswitch' #'sub-004_id-1122876_type-wtshelterswitch'

feature_array_path = experiment / 'feature_array.npy'
feature_array = load_npy_file(feature_array_path)

completed_distances_path = experiment / 'completed_distances.npy'
completed_distances = load_npy_file(completed_distances_path)

print('completed_distances shape:', completed_distances.shape)
print('compledted_distances:', completed_distances)

#plot the completed distances
plt.figure(figsize=(10, 5))
plt.plot(completed_distances, marker='o', linestyle='-', color='b')
plt.title('Completed Distances Over Time')
plt.xlabel('Time Step')
plt.ylabel('Distance')
plt.grid()
plt.show()
```
